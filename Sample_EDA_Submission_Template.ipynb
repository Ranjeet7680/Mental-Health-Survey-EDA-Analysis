## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# ================================
# Make Dataset Analysis Ready
# ================================

# 1. Create a copy of original data
df_clean = df.copy()

# 2. Drop duplicate rows
df_clean.drop_duplicates(inplace=True)

# 3. Handle Age outliers (keep reasonable human ages)
df_clean['Age'] = pd.to_numeric(df_clean['Age'], errors='coerce')
df_clean = df_clean[(df_clean['Age'] >= 18) & (df_clean['Age'] <= 70)]

# 4. Clean & standardize Gender column
df_clean['Gender'] = df_clean['Gender'].str.lower()

df_clean['Gender'] = df_clean['Gender'].replace({
    'male': 'Male', 'm': 'Male', 'man': 'Male', 'cis male': 'Male',
    'female': 'Female', 'f': 'Female', 'woman': 'Female', 'cis female': 'Female'
})

df_clean.loc[~df_clean['Gender'].isin(['Male', 'Female']), 'Gender'] = 'Other'

# 5. Fill missing categorical values with 'Unknown'
categorical_cols = df_clean.select_dtypes(include='object').columns
df_clean[categorical_cols] = df_clean[categorical_cols].fillna('Unknown')

# 6. Fill missing Age with median
df_clean['Age'].fillna(df_clean['Age'].median(), inplace=True)

# 7. Convert Yes/No columns to consistent format
yes_no_cols = [
    'self_employed','family_history','treatment','remote_work','tech_company',
    'benefits','care_options','wellness_program','seek_help','anonymity',
    'mental_health_consequence','phys_health_consequence','obs_consequence'
]

for col in yes_no_cols:
    if col in df_clean.columns:
        df_clean[col] = df_clean[col].str.capitalize()

# 8. Reset index
df_clean.reset_index(drop=True, inplace=True)

# 9. Final check
df_clean.info()
df_clean.head()
````

#### Chart - 14 - Correlation Heatmap

````python
# =========================================
# Correlation Heatmap (Categorical → Numeric)
# =========================================

# Step 1: Create a copy for correlation
df_corr = df_clean.copy()

# Step 2: Encode categorical variables numerically
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
for col in df_corr.select_dtypes(include='object').columns:
    df_corr[col] = le.fit_transform(df_corr[col])

# Step 3: Compute correlation matrix
corr_matrix = df_corr.corr()

# Step 4: Plot correlation heatmap
plt.figure(figsize=(16, 10))
sns.heatmap(
    corr_matrix,
    cmap='coolwarm',
    center=0,
    linewidths=0.5
)

plt.title("Correlation Heatmap of Mental Health Dataset")
plt.show()
````

##### 1. Why did you pick the specific chart?

I picked the Correlation Heatmap because the dataset contains many interrelated variables, and a heatmap is the most effective way to summarize and visualize relationships across all features at once.

This chart was chosen because:

It provides a global overview of how variables move together (positive, negative, or weak relationships).

It helps quickly identify strongly associated factors related to mental health outcomes such as treatment, work interference, and workplace support.

With many categorical variables encoded numerically, the heatmap allows pattern discovery that would be difficult through individual charts.

From a business and analytics perspective, it supports:

Feature selection for deeper analysis or modeling

Identifying key drivers of mental health behavior and perceptions

##### 2. What is/are the insight(s) found from the chart?

From the Correlation Heatmap, the following key insights can be derived:

Treatment shows noticeable positive correlations with variables such as family_history, benefits, seek_help, anonymity, and leave, indicating that both personal background and organizational support systems influence treatment-seeking behavior.

Variables related to workplace support (benefits, care_options, wellness_program, seek_help) are positively correlated with each other, suggesting that organizations offering one form of support are likely to provide others as well.

Mental health consequence, obs_consequence, and work_interfere exhibit associations, highlighting that perceived or observed stigma is linked with mental health impacting work and fear of negative outcomes.

Mental vs physical health perception correlates with multiple policy-related variables, implying that when companies offer strong support structures, employees are more likely to feel mental health is taken seriously.

Age shows weak correlations with most variables, suggesting that workplace culture and policies matter more than age in shaping mental health behavior.

#### Chart - 15 - Pair Plot

````python
# =========================================
# Pair Plot Visualization
# =========================================

# Select key variables for pair plot
pair_cols = [
    'Age',
    'treatment',
    'family_history',
    'work_interfere',
    'benefits'
]

# Create a copy and encode categorical variables
df_pair = df_clean[pair_cols].copy()

from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

for col in df_pair.select_dtypes(include='object').columns:
    df_pair[col] = le.fit_transform(df_pair[col])

# Plot pair plot
sns.pairplot(
    df_pair,
    diag_kind='kde'
)

plt.suptitle("Pair Plot of Key Mental Health Variables", y=1.02)
plt.show()
````

##### 1. Why did you pick the specific chart?

I picked the Pair Plot because it allows simultaneous visualization of multiple pairwise relationships between key variables related to mental health, instead of analyzing them one by one.

This chart was chosen because:

It shows how Age, treatment, family history, work interference, and benefits interact with each other in a single view.

The diagonal plots help understand the distribution of each variable, while off-diagonal plots reveal potential associations or patterns.

It is especially useful during EDA to identify clusters, trends, or separations that may not be visible in isolated charts.

From an analytics and business perspective, it helps narrow down important variable relationships worth deeper investigation.

##### 2. What is/are the insight(s) found from the chart?

From the Pair Plot, the following key insights can be observed:
